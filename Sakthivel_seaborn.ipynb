```python
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Set the theme to "dark grid"
sns.set_style("darkgrid")
```

```python
# Import the dataset
df = pd.read_csv("Diabetes.csv")
df.head()
```

```python
# Create scatter plot with Age and Insulin
sns.relplot(data=df, x="Age", y="Insulin")
plt.title('Scatter plot of Age vs Insulin')
plt.show()
```

```python
# Create scatter plot with Age and Insulin, and using Pregnancies for hue and size
sns.relplot(data=df, x="Age", y="Insulin", hue="Pregnancies", size="Pregnancies")
plt.title('Scatter plot of Age vs Insulin with hue and size by Pregnancies')
plt.show()
```

```python
# Create scatter plot with Age vs. Insulin, and use Pregnancies for hue and size
sns.relplot(data=df, x="Age", y="Insulin", hue="Pregnancies", size="Pregnancies")
plt.title('Scatter plot of Age vs Insulin with hue and size by Pregnancies')

# Set limits for the x and y axes to zoom in on the data
plt.xlim(20, 60)  # Setting limit for x-axis from 20 to 70
plt.ylim(-100, 100)   # Setting limit for y-axis from 0 to 400

plt.show()
```

The kind parameter in relplot can help you better understand the variable relationships in the figure. By default, relplot generates a scatter plot, but you can change the kind option to line or lineplot, scatter, swarm, and so on. This allows you to investigate the relationships between variables in a variety of ways. For example, using kind='line' may be useful for observing trends over continuous variables, whereas kind='swarm' may aid in visualizing the distribution of data points more clearly, particularly if there are overlapping points. As a result, adjusting the kind parameter based on the unique characteristics of your data can provide additional insights into the relationships between variables.

Using the col and row parameters in a relplot can also help you understand the variable relationships in the figure. These parameters enable you to generate distinct subplots based on the levels of one or more categorical variables. By dividing the data into smaller subsets based on categorical variables, you can more effectively compare the relationships between variables within each subset.

# Group 2

```python
# Load the dataset
salary_df = pd.read_csv("Salary Data.csv")
salary_df.columns
```

```python
salary_df.head()
```

```python
# Initial Exploration
#Scatter plot with Age vs Salary
sns.scatterplot(data=salary_df, x='Age', y='Salary')
plt.title('Age vs Salary')
plt.show()
```

```python
# Visualization
#Advanced Scatter plot with Gender and Education Level
plt.figure(figsize=(10, 6))
sns.scatterplot(data=salary_df, x='Age', y='Salary', hue='Gender', style='Education Level', palette='Set2')
plt.title('Age vs Salary with Gender and Education Level')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Styling and Customization
# Set the style to remove grid lines
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(data=salary_df, x='Age', y='Salary', hue='Gender', style='Education Level', palette='Set2')
plt.title('Age vs Salary with Gender and Education Level')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(False)
plt.show()
```

**Additional Analysis: Separate plots for each education level:**
Make distinct scatter plots using filtered data frames for every educational level.
To improve visualization, set the bounds of the x and y axes.

```python
# Additional Analysis
#Separate plots for each education level
education_levels = salary_df['Education Level'].unique()
for edu_level in education_levels:
    plt.figure(figsize=(8, 5))
    sns.scatterplot(data=salary_df[salary_df['Education Level'] == edu_level], x='Age', y='Salary', hue='Gender', palette='Set2')
    plt.title(f'Age vs Salary for {edu_level}')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(False)
    plt.ylim(bottom=0)  # Set y-axis limit to start from 0
    plt.xlim(left=0)    # Set x-axis limit to start from 0
    plt.show()
```

**Additional Analysis: Explore other Seaborn plot types:**
1. Create a box plot to visualize salary distribution across education levels by gender.
2. Create a bar plot to visualize the highest salary across education levels by gender.

```python
# Additional Analysis
#Explore other Seaborn plot types
plt.figure(figsize=(10, 6))
sns.boxplot(data=salary_df, x='Education Level', y='Salary', hue='Gender', palette='Set2')
plt.title('Salary distribution across Education Levels by Gender')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(False)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=salary_df, x='Education Level', y='Salary', hue='Gender', palette='Set2', estimator=max)
plt.title('Highest Salary across Education Levels by Gender')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(False)
plt.show()
```
