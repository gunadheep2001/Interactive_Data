# Question 2

A grouped bar chart with labels is a visual representation of data in which multiple bars are grouped together for each category and labeled to distinguish each bar within the groups. This type of chart is useful for comparing values from multiple categories and subcategories at once.

Here's how I would represent the type of data using a grouped bar chart with labels.

I'd like to show the distribution of household expenditures across various expense categories for three different income groups: low-income, middle-income, and high-income.

**Data structure**:\
**Categories**: Expense categories (for example, housing, transportation, and food).\
**Income groups**: low-income, middle-income, and high-income\
**Values**: Average expenditure in each expense category within each income group.

This chart will show bars for each expense category for each income group (Low-income, Middle-income, and High-income). Each bar in a group represents the average expenditure for that specific income category. Labels will be placed on each bar to indicate the expense category and its corresponding expenditure value.

This visualization allows for a comparison of spending patterns across income groups, highlighting potential disparities and areas where financial resources are distributed differently. It can help policymakers, researchers, and financial planners understand spending habits and make informed decisions about budgeting, policymaking, and financial advice.

# Question 5

```python
# Bar plot of the `Evaluation` variable
evaluation_counts = df['Evaluation'].value_counts()
fig, ax = plt.subplots(figsize=(8, 6))
evaluation_counts.plot(kind='bar', ax=ax)
ax.set_title('Evaluation')
ax.set_xlabel('Evaluation')
ax.set_ylabel('Count')

plt.show()
```
